```python
import os
import shutil


dataset_root = "../dataset_"  # change this to your dataset path

output_root = "data"
images_out = os.path.join(output_root, "images")
labels_out = os.path.join(output_root, "labels")


os.makedirs(images_out, exist_ok=True)
os.makedirs(labels_out, exist_ok=True)


for vid_folder in os.listdir(dataset_root):
    vid_path = os.path.join(dataset_root, vid_folder) # dataset/vid_*

    if not os.path.isdir(vid_path) or not vid_folder.startswith("vid_"):
        continue

    images_path = os.path.join(vid_path, "images/train") # dataset/vid_*/images/train
    labels_path = os.path.join(vid_path, "labels/train") # dataset/vid_*/labels/train

    if not os.path.exists(images_path) or not os.path.exists(labels_path):
        continue

    for img_file in os.listdir(images_path):
        if not img_file.endswith(".png"):
            continue

        old_img_path = os.path.join(images_path, img_file) # dataset/vid_*/images/train/frame_* 

        new_name = f"{vid_folder}_{img_file}"
        new_img_path = os.path.join(images_out, new_name)

        shutil.copy2(old_img_path, new_img_path)

        label_file = img_file.replace(".png", ".txt")
        old_label_path = os.path.join(labels_path, label_file)

        if os.path.exists(old_label_path):
            new_label_name = f"{vid_folder}_{label_file}"
            new_label_path = os.path.join(labels_out, new_label_name)

            shutil.copy2(old_label_path, new_label_path)
        else:
            print(f"⚠️ Missing label for {old_img_path}")



print("✅ Done combining dataset!")
```
