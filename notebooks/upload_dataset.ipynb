```python

from clearml import Dataset

DATASET_NAME  = "fatigue-region-labels-v2"
DATASET_PROJECT = "FatigueSense"

# Get latest version
parent = Dataset.get(
    dataset_name=DATASET_NAME,
    dataset_project=DATASET_PROJECT,
)

# Create new dataset version, with the previous version as parent
ds_new = Dataset.create(
    dataset_name=DATASET_NAME,
    dataset_project=DATASET_PROJECT,
    parent_datasets=[parent.id],
)

# UPLOAD NEW FILES HERE
# Note: Set the dataset_path to the name of your video. For the sake of convention, we will use the naming convention
# vid_XXX, where XXX is the three digit video number (e.g. vid_001, vid_002, etc.).
# CURRENT: vid_002, next vid_003. Update this line after updating.
# Upload the metadata files (data.yaml and train.txt) to the metadata folder, and the label files to the vid_XXX folder.

# Labels
ds_new.add_files(
    path="./dataset_2/labels",
    dataset_path="vid_002",
)
# Data.yaml
ds_new.add_files(
    path="./dataset_2/data.yaml",
    dataset_path="vid_002/metadata",
)
# Train.txt
ds_new.add_files(
    path="./dataset_2/train.txt",
    dataset_path="vid_002/metadata",
)

# Start upload
ds_new.upload(show_progress=True)
ds_new.finalize()
```
